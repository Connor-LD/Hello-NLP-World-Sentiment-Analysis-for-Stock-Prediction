## Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import pandas_datareader
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()

from utilities import *
```

## Initialize finBERT

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
```

## Get News Headline Data & Perform Sentiment Analysis

```python
### Using today's S&P500, the top 24 companies represent 41% of the index's performance.
### https://www.investopedia.com/ask/answers/08/find-stocks-in-sp500.asp

# # Small test
# stocks = ['AAPL', 'AMZN']

# Large test: top 24 stocks:
stocks = [
'AAPL', 'MSFT', 'AMZN', 'NVDA', 'GOOGL', 'GOOG', 'TSLA', 
'META', 'XOM', 'UNH', 'JNJ', 'JPM', 'V', 'PG', 'MA', 'CVX', 'HD', 
'ABBV', 'LLY','MRK', 'AVGO', 'PEP', 'KO', 'PFE']
```

```python
big_df = pd.DataFrame()
# Read finances and news for all stocks in list and create a big dataframe.
for stock in stocks:
    news = read_news(stock)
    finances = financial_dataset(stock)
    merged_df = merge_fin_news(finances, news)
    print(f"Performing sentiment analysis for {stock}")
    sentimentized_df = sentim_analyzer(merged_df, tokenizer, model)
    sentimentized_df = merge_dates(sentimentized_df)
    big_df = pd.concat([big_df, sentimentized_df], axis=0, ignore_index=True)
    big_df.to_csv('sp500.csv')
```

## EDA: Target Class

```python
incr = len(big_df[big_df['Price_change']==1])
decr = round(len(big_df[big_df['Price_change']!=1]),0)
incr_ratio = (100 * incr) / big_df.shape[0]
decr_ratio = (100 * decr) / big_df.shape[0]

fig, ax = plt.subplots(figsize=(5,5))
data = [incr, decr]
labels = ['Increases','Decreases']
plt.pie(data, labels=labels, shadow=True, autopct='%1.1f%%', startangle=90)
plt.title('Stock Fluctuations', fontsize=15)
plt.show()

print('{} ({:.2f}%) number of price increases \n \
{} ({:.2f}%) number of price decreases'
.format(incr, decr, incr_ratio, decr_ratio))
```
